//##IMAGE COMPRESSED WITH SVD##//

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

//get the jpg image//

```python
image=imread('nature.jpg')
if image.ndim==3:
    image_gray=image[:,:,0]+image[:,:,1]*0.5870+image[:,:,2]*0.1140
else:
    image_gray=image

u,s,vt=np.linalg.svd(image_gray,full_matrices=False)
k=int(0.1*len(s))
reconstructed_image=np.dot(u[:,:k],np.dot(np.diag(s[:k]),vt[:k,:]))
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(image_gray,cmap='gray')
plt.title("original image")
plt.subplot(1,2,2)
plt.imshow(reconstructed_image,cmap='gray')
plt.title(f"compressed image (top{k} components)")
plt.show()

   
```
